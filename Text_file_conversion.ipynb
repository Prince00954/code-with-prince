```python
def create_large_file(file_name, size_mb):
    """Creates a dummy text file of approximately size_mb in MB."""
    with open(file_name, 'w', encoding='utf-8') as f:
        line = "This is a sample line of text to fill the file.\n"
        while f.tell() < size_mb * 1024 * 1024:
            f.write(line)

# Example usage:
# create_large_file("input_200MB.txt", 200)
# create_large_file("input_400MB.txt", 400)
# You can generate other sizes similarly.

```

```python
import time

def convert_to_uppercase(input_file, output_file):
    """Converts the content of input_file to uppercase and writes to output_file."""
    start_time = time.time()

    with open(input_file, 'r', encoding='utf-8', errors='ignore') as infile, \
         open(output_file, 'w', encoding='utf-8') as outfile:
        for line in infile:
            outfile.write(line.upper())

    end_time = time.time()
    print(f"Time taken to convert {input_file}: {end_time - start_time:.2f} seconds")

```

```python
file_sizes = [200, 400, 600, 800, 1000]

# Create the input files
for size in file_sizes:
    input_file = f"input_{size}MB.txt"
    print(f"Creating {input_file}...")
    create_large_file(input_file, size)

# Convert the files to uppercase
for size in file_sizes:
    input_file = f"input_{size}MB.txt"
    output_file = f"output_upper_{size}MB.txt"
    print(f"\nConverting {input_file} to uppercase...")
    convert_to_uppercase(input_file, output_file)
```
